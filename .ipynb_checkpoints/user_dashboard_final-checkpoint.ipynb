# Dashboard

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.tsa.api import ExponentialSmoothing
import json
from flask import Flask, Response, request, jsonify, render_template

import warnings
warnings.filterwarnings("ignore")
```

### Import and arrange data

```python
forecasting = pd.read_json('./data/jewelry/jewelry_json.json')
# forecasting.drop(columns=['Unnamed: 0'], inplace=True)
forecasting['event_time'] = pd.to_datetime(forecasting['event_time'])
forecasting.set_index("event_time", inplace=True)
forecasting.sort_index(inplace=True)
forecasting.price.replace(0, 1, inplace=True)
forecasting = forecasting[forecasting.index.year.isin([2020])]
```

```python
forecasting_sales = pd.DataFrame(forecasting['price'].resample('W').sum())
forecasting_sales.price.replace(0, 1, inplace=True)
```

```python
train = forecasting_sales[:45]
test = forecasting_sales[45:]

#to view past train and test data:
# train['price'].plot()
# test['price'].plot(figsize=(10, 4), title='Weekly Sales');
```

### Forecasting Model - Weekly sales

```python
hw_price = ExponentialSmoothing(
    train['price'],
    seasonal_periods=7, #daily would be 24h if hourly data, 24*60 for minutes, 7 for day of week, etc
    trend='mul', #multiplicative
    seasonal='add'
).fit()

hw_predictions = hw_price.forecast(len(test))
```

```python
plt.figure(figsize=(12,6))
plt.plot(train['price'], label='Past Performance')
# plt.plot(test['price'], label='Test')
plt.plot(hw_predictions, label='Future Outlook')
plt.legend(loc='best')
plt.title('Weekly Total Sales and Future Outlook');
```

```python
def instructors(inst_num):
    instructor = forecasting[forecasting['instructor_id'] == inst_num]
    forecaster = pd.DataFrame(instructor['price'].resample('W').sum())
    forecaster.price.replace(0, 1, inplace=True)
    forecaster.to_json('./data/jewelry/forecaster_json.json')
    train = forecaster
    hw_price = ExponentialSmoothing(
        train['price'],
        seasonal_periods=7,
        trend='mul',
        seasonal='add'
    ).fit()
    hw_predictions = hw_price.forecast(6)
    hw_predictions.to_json('./data/jewelry/hw_predictions_update.json')
    
    plt.figure(figsize=(10,4))
    plt.plot(train['price'], label='Past Performance')
    plt.plot(hw_predictions, label='Future Outlook')
    plt.legend(loc='best')
    plt.title(f'Weekly Total Sales and Future Outlook for Instuctor {inst_num}');
      
    if 0.90*int(hw_predictions.mean()) <= int(train.tail(3).mean()) <= 1.10*int(hw_predictions.mean()):
        print('There may be a similar amount of lessons in the next 3 months!')
    else:
        if int(train.tail().mean()) > int(hw_predictions.mean()):
            print('There may be fewer lessons in the next 3 months!')
        if int(train.tail().mean()) < int(hw_predictions.mean()):
            print('Amount of lessons may be increasing in the next 3 months!')
```
